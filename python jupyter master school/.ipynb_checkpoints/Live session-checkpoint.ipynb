# A mystery in Bunnyville 

In a quaint little town nestled in the picturesque countryside, there's a peculiar phenomenon unfolding. 
The townspeople live in harmony with the local population of bunnies, but recently, the relationships between the two have taken a dark turn. 
These innocent-looking bunnies have been meeting with the unsuspecting residents, and in some cases, the encounters have ended in tragedy.


You've been hired by the mayor to explore these mysterious events and figure out the underlying causes. 
Your mission is to analyze the meetings between the townspeople and the bunnies, identify patterns, and determine the factors that might lead to the murders occurring in these encounters.



Delve into this peculiar situation and report your findings to the mayor.

Your recommendations will be crucial in determining the future of the town and the safety of its residents.

## What is SQLalchemy?


SQLAlchemy is a Python library for working with relational databases. It provides a set of high-level abstractions for interacting with databases, allowing developers to write Python code that can be used across multiple database systems. SQLAlchemy supports a wide range of database systems, including PostgreSQL, MySQL, SQLite, and Oracle.









## What is BeautifulSoup?

BeautifulSoup is a Python library used for web scraping purposes. It allows developers to parse HTML and XML documents and extract data from them. BeautifulSoup provides a simple and intuitive way to navigate and search through the document tree using Pythonic idioms, making it a popular choice for web scraping tasks.
